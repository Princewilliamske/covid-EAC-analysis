```python
import pandas as pd

df = pd.read_csv("owid-covid-data.csv")
df.columns  # Display column names
df.head()   # Preview first few rows
df.isnull().sum()  # Check for missing data
```

```python
df['date'] = pd.to_datetime(df['date'])
df.fillna(method='ffill', inplace=True)
```

```python
#Total Cases Over time
import matplotlib.pyplot as plt
import seaborn as sns

selected_countries = df[df['location'].isin(['Tanzania', 'Uganda', 'Kenya', 'Rwanda', 'Burundi', 'South Sudan', 'Somalia', 'Djibouti', 'Ethiopia', 'Congo'])]
sns.lineplot(data=selected_countries, x='date', y='total_cases', hue='location')
plt.title('Total Cases Over time')
```

```python
#Total Deaths over time
sns.lineplot(data=selected_countries, x='date', y='total_deaths', hue='location')
plt.title('Total Deaths Over Time')
```

```python
#Daily New Cases Comparison
sns.lineplot(data=selected_countries, x='date', y='new_cases', hue='location')
plt.title('Daily new Cases Comparison')
```

```python
#Death Rate calculation
import numpy as np

selected_countries['death_rate'] = np.where(
    selected_countries['total_cases'] > 0,
    selected_countries['total_deaths'] / selected_countries['total_cases'],
    0
)

plt.figure(figsize=(12, 6))
sns.lineplot(data=selected_countries, x='date', y='death_rate', hue='location')
plt.title('COVID-19 Death Rate Over Time')
plt.ylabel('Death Rate (deaths / cases)')
plt.xlabel('Date')

plt.show()
```

```python
#Total Vaccinations Over Time
sns.lineplot(data=selected_countries, x='date', y='total_vaccinations', hue='location')
plt.title('Total Vaccinations Over Time')
```

```python
#Percent of Vaccinated Population
sns.lineplot(data=selected_countries, x='date', y='people_vaccinated_per_hundred', hue='location')
plt.title('Percent of Vaccinated Population')
```
